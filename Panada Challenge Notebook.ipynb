```python
# Dependencies and Setup
import pandas as pd

# File to Load 
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete_df.head()
```

```python
#Calculate the Total Number of Schools
Total_School_Count = school_data_complete_df['school_name'].nunique()
Total_School_Count
```

```python
#Calculate the total number of students
Total_Students = school_data_complete_df["student_name"].count()
Total_Students
```

```python
#Calculate the total budget
Total_Budget = school_data_complete_df["budget"].unique().sum()
Total_Budget
```

```python
#Calculate the average maths score
Ave_Math_Score = school_data_complete_df["maths_score"].mean()
Ave_Math_Score
```

```python
#Calculate the average reading score
Ave_Reading_Score = school_data_complete_df["reading_score"].mean()
Ave_Reading_Score
```

```python
#Calculate the percentage of students with a passing maths score (50 or greater)
Passing_math = school_data_complete_df[school_data_complete_df['maths_score']>=50]
Passing_math_percentage = len(Passing_math)/Total_Students*100
Passing_math_percentage
```

```python
#Calculate the percentage of students who passed maths and reading (% Overall Passing)
Passing_reading = school_data_complete_df[school_data_complete_df['reading_score']>=50]
Passing_reading_percentage = len(Passing_reading)/Total_Students*100
Passing_reading_percentage
```

```python
#Calculate the percentage of students who passed maths and reading (% Overall Passing)
Overall_passing = school_data_complete_df[(school_data_complete_df['maths_score']>=50) & (school_data_complete_df['reading_score']>=50)]
Overall_passing_percentage = len(Overall_passing)/Total_Students*100
Overall_passing_percentage 
```

```python
#Create a dataframe to hold the above results
Summary_Data = pd.DataFrame({'Total Schools': Total_School_Count,
                               'Total Students': Total_Students, 
                               'Total Budget': Total_Budget, 
                               'Average Math Score': Ave_Math_Score,
                               'Average Reading Score': Ave_Reading_Score,
                               '% Passing Math': Passing_math_percentage,
                               '% Passing Reading': Passing_reading_percentage,
                               '% Overall Passing':Overall_passing_percentage
                                }, index=[0]) 
```

```python
Summary_Data.head()
```

```python
#Optional: give the displayed data cleaner formatting
Summary_Data['Total Students'] = Summary_Data ['Total Students'].map('{:,}'.format)
Summary_Data['Total Budget'] = Summary_Data ['Total Budget'].map('${:,.2f}'.format)
```

```python
# Cleaner formatting for the rest of the results by using mapping (optional)
Summary_Data['Average Math Score'] = Summary_Data['Average Math Score'].map('{:.2f}'.format)
Summary_Data['Average Reading Score'] = Summary_Data['Average Reading Score'].map('{:.2f}'.format)
Summary_Data['% Passing Math'] = Summary_Data['% Passing Math'].map('{:.2f}%'.format)
Summary_Data['% Passing Reading'] = Summary_Data['% Passing Reading'].map('{:.2f}%'.format)
Summary_Data['% Overall Passing'] = Summary_Data['% Overall Passing'].map('{:.2f}%'.format)
```

```python
# School Summary
school_data_complete_df = school_data_complete_df.set_index('school_name')
Grouped_School_Name = school_data_complete_df.groupby(["school_name"])
```

```python
# Average reading score
Ave_Reading_Score = Grouped_School_Name["reading_score"].mean()
Ave_Reading_Score 
```

```python
# Average math score
Ave_Math_Score = Grouped_School_Name["maths_score"].mean()
Ave_Math_Score
```

```python
# Total students
Total_Students = Grouped_School_Name.size()
Total_Students
```

```python
# Per student budget
Student_Budget = Grouped_School_Name['budget'].mean()/Grouped_School_Name['size'].mean()
Student_Budget
```

```python
# Total budget
Total_Budget = Total_Students * Student_Budget
Total_Budget
```

```python
# Percentage of students who passed math
Passed_Math = school_data_complete_df.loc[school_data_complete_df["maths_score"] >= 50].groupby(["school_name"]).size()
Percentage_Math_Passed = Passed_Math/Total_Students*100
```

```python
# Percentage of students who passed reading
Passed_Reading = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 50].groupby(["school_name"]).size()
Percentage_Reading_Passed = Passed_Reading/Total_Students*100
```

```python
# Percentage of students who passed math AND reading
Passed_Overall = school_data_complete_df.loc[(school_data_complete_df["maths_score"] >= 50) & (school_data_complete_df["reading_score"]>=50)].groupby(['school_name']).size()
Percentage_passed_overall = Passed_Overall/Total_Students*100
```

```python
# Created a data frame to hold above results
School_Summary = pd.DataFrame({'School Type': Grouped_School_Name['type'].unique(),
                               'Total Students': Total_Students,
                               'Total School Budget': Total_Budget,
                               'Per Student Budget': Student_Budget, 
                               'Average Math Score': Ave_Math_Score, 
                               'Average Reading Score': Ave_Reading_Score,
                               '% Passing Math': Percentage_Math_Passed,
                               '% Passing Reading': Percentage_Reading_Passed,
                               '% Overall Passing':Percentage_passed_overall                         
                               })
```

```python
School_Summary.head()
```

```python
# Used .apply(lambda) to change school type from an array
School_Summary['School Type'] = School_Summary['School Type'].apply(lambda t: t[0])
```
